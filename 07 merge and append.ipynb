```python
cost = pd.DataFrame({
    "name" : ["A", "B", "C"],
    "CP" : [50, 100, 200]
})
```

```python
cost
```

```python
sell = pd.DataFrame({
    "name" : ["A", "B", "D"],
    "SP" : [100, 200, 300],
})
```

```python
sell
```

```python

pd.merge(cost, sell, on= ['name', 'name'], how='inner')
```

```python
pd.merge(cost, sell, left_on=['CP'], right_on=['SP'])
```
